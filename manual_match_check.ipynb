```python
import pandas as pd
from pathlib import Path

src = Path("/Users/stefan/Desktop/Thesis/v4/Companies House Data/uk_comps_with_ch_data_funded.csv")
out = src.with_name("uk_comps_with_ch_data_matched_sample_100.csv")

df = pd.read_csv(src, low_memory=False)
sampled = (
    df[df["ch_match_reason"] == "matched"]
      .sample(n=200, random_state=42)  # reproducible sample
)

sampled.to_csv(out, index=False)
print(f"Saved {len(sampled):,} rows to {out}")
sampled.head()
```
